```python
import matplotlib.pyplot as plt

# Extract the necessary columns
x = product_sheet['Date Add']
y = product_sheet['Price']

# Create the plot
plt.plot(x, y, marker='o', linestyle=' ', color='blue')

# Customize the plot
plt.title('Product Price Over Time')
plt.xlabel('Time')
plt.ylabel('Price')
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Filter out records with date smaller than '2015-01-01'
filtered_data = product_sheet.loc[product_sheet['Date Add'] >= '2015-01-01']

# Extract the necessary columns from the filtered data
x = filtered_data['Date Add']
y = filtered_data['Price']

# Create the plot
plt.plot(x, y, marker='o', linestyle=' ', color='blue')

# Customize the plot
plt.title('Product Price Over Time (Filtered)')
plt.xlabel('Time')
plt.ylabel('Price($)')
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Calculate the total purchase amount of each product ID
total_purchase_amount = transaction_sheet.groupby('Product ID')['Purchase Amount'].sum()
total_purchase_amount = total_purchase_amount.reset_index()
# Print the total purchase amount
total_purchase_amount.head()
```

```python
# Merge transaction_sheet and product_sheet based on a common column
merged_data = pd.merge(filtered_data, total_purchase_amount, on='Product ID')

merged_data.head()
```

```python
# Extract the necessary columns from the filtered data
x = merged_data['Purchase Amount']
y = merged_data['Price']

# Create the plot
plt.plot(x, y, marker='o', linestyle=' ', color='blue')

# Customize the plot
plt.title('Product Price and Total Product Sale')
plt.xlabel('Product Sale')
plt.ylabel('Price($)')
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Extract the necessary columns from the filtered data
x = merged_data['Date Add']
y = merged_data['Purchase Amount']

# Create the plot
plt.plot(x, y, marker='o', linestyle=' ', color='blue')

# Customize the plot
plt.title('Total Product Sale Over Time')
plt.xlabel('Time')
plt.ylabel('Total Product Sale')
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Calculate the total purchase amount of each product ID
total_customer_purchase_amount = transaction_sheet.groupby('Customer ID')['Purchase Amount'].sum()
total_customer_purchase_amount = total_customer_purchase_amount.reset_index()
# Print the total purchase amount
total_customer_purchase_amount.head()
```

```python
merged_user_transaction_data = pd.merge(customers_sheet, total_customer_purchase_amount, on="Customer ID")
merged_user_transaction_data.drop('Full Name', axis = 1, inplace = True)

merged_user_transaction_data.head()
country_mapping = {'VN': 1, 'CN': 2, 'USA': 3, 'UK': 4, 'ES': 5, 'FR': 6, 'RU': 7, 'JP': 8}

merged_user_transaction_data['Country_Num'] = merged_user_transaction_data['Country'].map(country_mapping)
merged_user_transaction_data.drop('Country', axis = 1, inplace = True)
merged_user_transaction_data.head()
```

```python
y = merged_user_transaction_data['Purchase Amount']
x = merged_user_transaction_data['DOB']

plt.plot(x, y, marker='o', linestyle=' ', color='blue')

plt.title('DOB and Purchase Amount')
plt.xlabel('DOB')
plt.ylabel('Purchase Amount')
plt.xticks(rotation=45)

plt.show()
```
